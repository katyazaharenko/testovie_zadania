```python
import pandas as pd
import sqlite3
```

```python
data = pd.read_excel("Table1.xlsx", header=2)
```

```python
conn = sqlite3.connect("test_ADV.db")
cursor = conn.cursor()
```

```python
data.to_sql("Table1", conn, index=False)
```

## #1


```python
sql_text = """SELECT 
            [Months], 
            [Advertisers list], 
            [Brands list], 
            [TVR] 
            FROM Table1"""

cursor.execute(sql_text)
res_1 = cursor.fetchall()
res_1 = pd.DataFrame(res_1)
res_1.columns = ["Months", "Advertisers list", "Brands list", "TVR"]
res_1
```

Производится вывод таблицы «Table1», из всех столбцов выбираются «Months», «Advertisers List», «Brands list», «TVR», то есть просто фильтруем таблицу по столбцам.

Смысл: из данных по рекламным роликам выводим значение месяца трансляции ролика на ТВ, список рекладомателей данного ролика, список обнаруженных брендов, и ТВ-рейтинг ролика как отношение количества человек, увидевшего ролик за среднюю минуту, к генеральной совокупности (население России старше 4 лет).

## #2

```python
sql_text = """SELECT 
            [Months], 
            [Advertisers list], 
            [Brands list], 
            [TVR] 
            FROM Table1 
            WHERE [Brands list] = 'ARBIDOL' """

cursor.execute(sql_text)
res_2 = cursor.fetchall()
res_2 = pd.DataFrame(res_2)
res_2.columns = ["Months", "Advertisers list", "Brands list", "TVR"]
res_2
```

Из всех столбцов таблицы "Table1" выбираются «Months», «Advertisers List», «Brands list», «TVR», данные фильтруются по столбцу «Brands list» = 'ARBIDOL'.

Смысл: то же самое, что в первом запросе, только в данном нас интересуют рекламные ролики, где был только бренд ARBIDOL.

## #3

```python
sql_text = """SELECT 
            [Months], 
            [Advertisers list], 
            [Brands list], 
            IIF([Brands list] like '%NICORETTE%' AND [National channel] = 'STS', [TVR]*1.5, [TVR]) as AdjTVR 
            FROM Table1 """

cursor.execute(sql_text)
res_3 = cursor.fetchall()
res_3 = pd.DataFrame(res_3)
res_3.columns = ["Months", "Advertisers list", "Brands list", "AdjTVR"]
res_3
```

Производится вывод таблицы «Table1», из всех столбцов выбираются «Months», «Advertisers List», «Brands list», и для тех строк, где в столбце "Brands list" есть подстрока 'NICORETTE' с любым количеством символов справа и слева и значение столбца "National channel" равно STS, значение столбца "AdjTVR" равно 1.5 * «TVR», иначе - «TVR».

Смысл: то же самое, что в первом запросе, но для рекламных роликов, включающих бренд 'NICORETTE' (в качестве отдельного наименования или части другого слова), и транслирующихся на СТС, значение рейтинга увеличивается на 50%.

## #4

```python
sql_text = """SELECT 
            [Months], 
            [Advertisers list], 
            [Brands list], 
            SUM([TVR]) as TVR 
            FROM Table1 
            GROUP BY [Months], [Advertisers list], [Brands list]"""

cursor.execute(sql_text)
res_4 = cursor.fetchall()
res_4 = pd.DataFrame(res_4)
res_4.columns = ["Months", "Advertisers list", "Brands list", "TVR"]
res_4
```

Данные таблицы "Table1" группируются по столбцам «Months», «Advertisers List», «Brands list», по этим группам считается агрегат - суммарный TVR.

Смысл: считаем суммарный рейтинг по каждой тройке "месяц-рекламодатели-бренды".

P.S. для корректного запуска убрала круглые скобки в части `GROUP BY ([Months], [Advertisers list], [Brands list])`. Возможно, работает на старых версиях SQL Server.

## Выгрузка результатов
