Заметки по сводке по уникальным значениям в колонках данных:
- order_online_payment_flg - только одно значение, везде онлайн-оплата, а какая еще бывает (наличные как будто не применимы, перевод?)
- account_rk - у счета `9b2d5b4678781e53038e91ea5324530a03f27dc1d0e5f6c9bc9d493a23be9de0` подозрительно много записей (8%)
- order_rk - по заказу `9b2d5b4678781e53038e91ea5324530a03f27dc1d0e5f6c9bc9d493a23be9de0` подозрительно много записей (6%)
- bundle_nm - `Premium` и `Tinkoff Premium`, `Pro` и `Tinkoff Pro` - полагаю, что это названия одних и тех же сервисов после и до ребрендинга
- order_status_cd - `Статус заказа (расшифровка внизу)`, не нашла расшифровку в описании датасета, скорее всего, `SUC` = Successful
- party_first_order_dt - в топе дат первых заказов - праздники (8 марта и 23 февраля). Скорее всего, была какая-то акция (скидка на первый заказ по промокоду, например)
- hotel_country - бОльшая часть бронирований (14%) приходится на российские отели
- hotel_city - бОльшая часть бронирований (2%) относится к московским отелям
- avia_dep_city и avia_arr_city - больше всего перелетов из/в Москву
- loyalty_accrual_rub_amt - есть отрицательные значения в начислении баллов лояльности, не понятно, почему так бывает
- bounce_cd - в 3% случаев код 5 - ошибка при отправке письма
- last_sms_success_flg - уточнить, есть ли связь между заказами и успешным СМС
- email_valid_flg - уточнить, есть ли связь между заказами и валидными email-ами
- delivery_region_category_cd - расшифровка типа населенного пункта?
- lvn_city_nm - 3% клиента живит в Екатеринбурге, при этом большая часть клиентов получали карту в Москве. Скорее всего Москва не указывается как город, а указывается как lvn_state_nm
- time_zone_cd - есть таймзоны разного формата `^%d{1}%d{1}$` и `^0(%d|0){1}%d{1}$`. В рамках анализа можно считать, что, например, `003` и `3` - это одна таймзона
- last_used_product_cd и first_used_product_cd - коды продуктов без расшифровки
- marital_status_cd - есть ли зависимость между покупкой билетов и бронированием отелей и семейным статусом? Возможно, стоит предлагать замужним скидку на второй заказ по тому же направлению
